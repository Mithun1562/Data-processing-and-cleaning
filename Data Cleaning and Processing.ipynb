```python
#Importing necessary packages needed to clean and manupilate the data
import numpy as np
import pandas as pd
```

```python
#Importing and loading the playstore raw dataset using pandas read_csv() to play_store variable
play_store = pd.read_csv('C:\\Users\\Mithunsomu\\Downloads\\googleplaystore.csv',sep=',')
play_store.head()
```

```python
#Creating a boolean masking of null values. Here False refers to the data that are not null and True for Null values
play_store.isnull()
```

```python
#Checking the number of null values in columns wise using .isnull() which creates boolean Frame and .sum() summerizes the True Null values.
#Here we can see that Ratings column has lot of null values
play_store.isnull().sum()
```

```python
#Retrieving all the rows that has Null values in the Ratings column
#Here we got the entire data that has null's
play_store[play_store['Rating'].isnull()]
```

```python
#Removing the rows that has Null values in the Rating column 
#Here the row that has atleast one null value also gets removed
play_store=play_store[~play_store.Rating.isnull()]
play_store
```

```python
#Null values in the Ratings column is now removed
#Below we can see now the null values are 0 in Rating column
play_store.isnull().sum()
```

```python
#Retrieving all the rows that has Null in the Content Ratings column
#Here we can see that only one row has null value from content rating column
play_store[play_store['Content Rating'].isnull()]
```

```python
#Removing the rows that has Null values in the Content Rating column 
#Here one row from content rating get's deleted
play_store=play_store[~play_store['Content Rating'].isnull()]
play_store
```

```python
#Removing the rows that has Null values in the Current Ver column 
#Here 4 rows containing null values gets deleted
play_store=play_store[~play_store['Current Ver'].isnull()]
play_store
```

```python
#Removing the rows that has Null values in the Android Ver column 
#Here 3 rows containing null vales get's removed
play_store=play_store[~play_store['Android Ver'].isnull()]
play_store
```

```python
#No null value is remaining now
#Below we can now see that there are no null's present in out dataset
play_store.isnull().sum()
```

```python
#Here we are checking to confirm if any null vales are missed using np.NAN  from NumPy
(play_store.loc[:]==np.NAN).sum()
```

```python
#Removing the '+' sign from the Installs column inorder to change it's data type into int64

play_store['Installs']=play_store['Installs'].replace({'\+':''},regex=True)
play_store
```

```python
#Removing the '+' and ',' sign from Installs column
play_store['Installs'] = play_store['Installs'].replace({'\+':'',',':''},regex=True)
play_store
```

```python
#By using .info() method we can easily get some basic info about the data such as range , data type and set data type accourding to it
#Below we can see that Reviews column has numeric data but the data is set to object data type, so we have to change it's datatype accourdingly
play_store.info()
```

```python
#Here we are changing the datatype of Reviews column from Object to int64 to continue with numerical analysis.
#Below now we can see that ther reviews column dtype has got changed from object to int64 bit.
play_store['Reviews'] = play_store['Reviews'].astype('int64')
play_store.info()
```

```python
# After removing the '+' and ',' signs from Installs column now we are changing the datatype of Installs column from Object to int64 to continue with numerical analysis.
#Below now we can see that ther Installs column dtype has got changed from object to int64 bit.
play_store['Installs'] = play_store['Installs'].astype('int64')
play_store.info()
```

```python
#Here we are changing the datatype of Last updated column from Object to datetime as it has timeseries data's to continue with numerical analysis.
#Below now we can see that ther Last Updated column dtype has got changed from object to datetime.
play_store['Last Updated'] = pd.to_datetime(play_store['Last Updated'])
play_store.info()
```

```python
#Importing and loading the playstore raw dataset using pandas read_csv() to app_store variable.
app_store = pd.read_csv('C:\\Users\\Mithunsomu\\Downloads\\AppleStore.csv',sep=',',index_col=0)
app_store.head()
```

```python
#By using .info() method we can easily get some basic info about the data such as range , data type and set data type accourding to it

app_store.info()
```

```python
#Checking if there are any remaning null values
#Below we can see that there are no null values in the dataset
app_store.isnull().sum()
```

```python
#Checking if the column 'track_name' has all ASCII values
app_store['track_name'].map(lambda x:x.isascii()).value_counts()
```

```python
#Removing the track_name having Non English Values

app_store['track_name'] = app_store['track_name'].apply(lambda row: row.encode('ascii',errors='ignore').decode())

app_store['track_name']=app_store['track_name'].replace(['-','--'],'Random App')
```

```python
#Removing '+' sign from the column to change it's type to int
app_store['cont_rating']=app_store['cont_rating'].replace({'\+':''},regex=True)
```

```python
#Changing the Data type from object to int to 
app_store['cont_rating']=app_store['cont_rating'].astype('int64')
app_store.info()
```

```python
#From below code we can easily know that there anr no null value present in the app store dataset and the datatpyes are also set accourdingly
(app_store.loc[:]==np.NAN).sum()
```
